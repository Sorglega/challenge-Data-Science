#1. Análisis de facturación



```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos tienda
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Lugar de Compra": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Precio": [164300.0, 192300.0, 209600.0, 757500.0, 335200.0],
    "Costo de envío": [6900.0, 8400.0, 15900.0, 41000.0, 20200.0],
    "Calificación": [4, 1, 1, 4, 5],
    "lat": [4.60971, 6.25184, 10.39972, 3.43722, 6.25184],
    "lon": [-74.08175, -75.56359, -75.51444, -76.52250, -75.56359]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Calcular facturación total por lugar
facturacion_por_lugar = df.groupby("Lugar de Compra")["Precio"].sum()

# Graficar facturación por lugar
plt.figure(figsize=(8, 6))
facturacion_por_lugar.plot(kind="bar", color="skyblue")
plt.title("Facturación por Lugar de Compra")
plt.xlabel("Lugar de Compra")
plt.ylabel("Facturación Total")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos tienda
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Categoría del Producto": ["Electrónicos", "Muebles", "Juguetes", "Electrodomésticos", "Muebles"],
    "Lugar de Compra": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Precio": [164300.0, 192300.0, 209600.0, 757500.0, 335200.0],
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Agrupar las ventas por categoría y tienda
ventas_por_categoria_tienda = df.groupby(["Lugar de Compra", "Categoría del Producto"])["Precio"].sum().unstack()

# Visualizar los datos
plt.figure(figsize=(10, 6))
ventas_por_categoria_tienda.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="viridis")
plt.title("Ventas por Categoría Según Tienda")
plt.xlabel("Lugar de Compra")
plt.ylabel("Ventas Totales")
plt.xticks(rotation=45)
plt.legend(title="Categoría del Producto")
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos tienda
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Lugar de Compra": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Calificación": [4, 1, 1, 4, 5],
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Calcular calificación promedio por tienda
calificacion_promedio = df.groupby("Lugar de Compra")["Calificación"].mean()

# Graficar calificación promedio por tienda en un gráfico circular
plt.figure(figsize=(8, 6))
calificacion_promedio.plot(kind="pie", autopct="%1.1f%%", startangle=90, colors=["lightblue", "lightgreen", "orange", "pink"])
plt.title("Distribución de Calificación Promedio por Tienda")
plt.ylabel("")  # Quitar etiqueta del eje Y
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

# Datos tienda
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Lugar de Compra": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Precio": [164300.0, 192300.0, 209600.0, 757500.0, 335200.0],
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Calcular el producto más y menos vendido por tienda
productos_por_tienda = df.groupby("Lugar de Compra")["Precio"]
mas_vendidos = productos_por_tienda.idxmax().apply(lambda idx: df.loc[idx, "Producto"])
menos_vendidos = productos_por_tienda.idxmin().apply(lambda idx: df.loc[idx, "Producto"])

# Crear un DataFrame resumen
resumen = pd.DataFrame({
    "Producto Más Vendido": mas_vendidos,
    "Producto Menos Vendido": menos_vendidos
})

print(resumen)
```

# 5. Envío promedio por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos tienda
data = {
    "Producto": ["Asistente virtual", "Mesa de comedor", "Juego de mesa", "Microondas", "Silla de oficina"],
    "Lugar de Compra": ["Bogotá", "Medellín", "Cartagena", "Cali", "Medellín"],
    "Costo de envío": [6900.0, 8400.0, 15900.0, 41000.0, 20200.0],
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Calcular el costo de envío promedio por tienda
envio_promedio_por_tienda = df.groupby("Lugar de Compra")["Costo de envío"].mean()

# Graficar el envío promedio por tienda
plt.figure(figsize=(8, 6))
envio_promedio_por_tienda.sort_values(ascending=False).plot(
    kind="barh", color="lightblue", edgecolor="black"
)
plt.title("Costo de Envío Promedio por Tienda (Depresión Visual)")
plt.xlabel("Costo de Envío Promedio")
plt.ylabel("Lugar de Compra")
plt.tight_layout()
plt.show()
```
